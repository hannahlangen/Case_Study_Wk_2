```python
def clean_gender(x):
    if x in ['M', 'Male']:
        return 'Male'
    elif x in ['F', 'Femal','female']:
        return 'Female'
    else:
        return 'U'
```

```python
def clean_states(x):
    if x in ['AZ', 'Arizona']:
        return 'Arizona'
    if x in ['Nevada']:
        return 'Nevada'
    if x in ['Oregon']:
        return 'Oregon'
    if x in ['California','Cali']:
        return 'California'
    elif x in ['WA', 'Washington']:
        return 'Washington'
```

```python
cs_df['State']=list(map(clean_states,cs_df['State']))
```
