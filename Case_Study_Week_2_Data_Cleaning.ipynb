```python
def load_original_data():
    file1 = pd.read_csv('Data copy/file1.csv')
    file2 = pd.read_csv('Data copy/file2.csv')
    file3 = pd.read_csv('Data copy/file3.csv')
    return pd.concat([file1,file2,file3], axis=0)
```

```python
#drop "Customer" column
def drop_columns(cs_df) :
    cs_df.drop(columns=["Customer"], inplace=True)
    return cs_df
drop_columns(cs_df)
```

```python
#use function to clean 'Customer Lifetime Value' - remove "%"
def clean_cust_lifetime_value(value):
    if "%" in str(value):
        return float(value.replace("%",""))/100
    else:
        return float(value)
```

```python
cs_df['Customer Lifetime Value']=list(map(clean_cust_lifetime_value,cs_df['Customer Lifetime Value']))
```

```python
#use function to clean "Gender"
def clean_gender(x):
    if x in ['M', 'Male']:
        return 'Male'
    elif x in ['F', 'Femal','female']:
        return 'Female'
    else:
        return 'U'
```

```python
cs_df['Gender']=list(map(clean_gender,cs_df['Gender']))
```

```python
def clean_states(x):
    if x in ['AZ', 'Arizona']:
        return 'Arizona'
    if x in ['Nevada']:
        return 'Nevada'
    if x in ['Oregon']:
        return 'Oregon'
    if x in ['California','Cali']:
        return 'California'
    elif x in ['WA', 'Washington']:
        return 'Washington'
```

```python
cs_df['State']=list(map(clean_states,cs_df['State']))
```

```python
#clean Number of Open Complaints - only return the third digit
def return_third_digit(value):
    value = str(value)
    if "/" in str(value):
        return value.split("/")[1]
    else:
        return value
```

```python
cs_df['Number of Open Complaints']=list(map(return_third_digit, cs_df['Number of Open Complaints']))
```

```python
#sort states into regions
def state_region_labels(x):
    if x in ['California']:
        return 'West'
    if x in ['Oregon']:
        return 'North West'
    if x in ['Washington']:
        return 'East'
    elif x in ['Nevada','Arizona']:
        return 'Central'
```

```python
cs_df['State']=list(map(state_region_labels,cs_df['State']))
```
